```python
betfirst = ipl[ipl['win_by_runs'] != 0]
```

```python
betfirst.head()
```

```python
plt.figure(figsize=(4,5))
plt.hist(betfirst['win_by_runs'])
plt.title('Distribution of Run')
plt.xlabel("Runs")
plt.show()
```

```python
betfirst['winner'].value_counts()
```

```python
plt.figure(figsize=(6,5))
plt.bar(
    list(betfirst['winner'].value_counts()[0:3].keys()),
    list(betfirst['winner'].value_counts()[0:3]),
    color = ['gray', 'red' , 'blue'])
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.pie(
    list(betfirst['winner'].value_counts()),
    labels = list(betfirst['winner'].value_counts().keys()),
    autopct='%0.1f%%')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.pie(
    list(betfirst['winner'].value_counts()[0:6]),
    labels = list(betfirst['winner'].value_counts()[0:6].keys()),
    autopct='%0.1f%%')
plt.show()
```

```python
secondbet = ipl[ipl['win_by_wickets']!=0]
```

```python
secondbet.head()
```

```python
plt.figure(figsize = (5,5))
plt.hist(
    list(secondbet['win_by_wickets'].value_counts()),
    color = 'olive',
    bins =30)
plt.show()
```

```python
secondbet['winner'].value_counts()
```

```python
plt.figure(figsize = (6,5))
plt.bar(
    list(secondbet['winner'].value_counts()[0:3].keys()),
    list(secondbet['winner'].value_counts()[0:3]),
    color = ['olive' , 'purple' , 'orange'])
plt.show()
```

```python
plt.figure(figsize = (4,4))
plt.pie(
    list(secondbet['winner'].value_counts()[1:6]),
    labels= list(secondbet['winner'].value_counts()[1:6].keys()),
    autopct = '%0.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
categories = list(secondbet["win_by_wickets"].value_counts()[1:6].keys())
values =list(secondbet['win_by_wickets'].value_counts()[1:6])
colors = sns.color_palette("rainbow", len(categories))
sns.barplot(
     x=categories,
     y =values,
     palette=colors)
plt.show()
```
